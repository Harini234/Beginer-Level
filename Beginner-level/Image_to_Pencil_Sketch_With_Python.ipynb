Importing Standard Libraries

```python
from google.colab.patches import cv2_imshow
```

```python
import cv2
```

Reading the image.

```python
img = cv2.imread('/content/drive/MyDrive/LGM_VIP/rose.jpg')
```

Displaying the image.

```python
cv2_imshow(img)
```

Making the image in RGB format

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
cv2_imshow(img_rgb)
```

Converting RGB format into gray scale.

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2_imshow(img_gray)
```

Inverting the gray scale image into negative image.

```python
img_invert = cv2.bitwise_not(img_gray)
```

```python
cv2_imshow(img_invert)
```

Making the image blurry by using cv2.GaussianBlur function.

```python
img_smoothing = cv2.GaussianBlur(img_invert,(21,21),sigmaX=0,sigmaY=0)
cv2_imshow(img_smoothing)
```

Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

```python
def dodgeV2(x,y):
    return cv2.divide(x,255-y,scale=256)
```

```python
final_img = dodgeV2(img_gray,img_smoothing)
cv2_imshow(final_img)
```
